## Listas Encadeadas

#### Desvantagens de Vetores

* Em um vetor não ordenado, a busca é lenta
* Em um vetor ordenado, a inserção é lenta
* Em ambos, a remoção é lenta
* O tamanho do vetor "não pode" ser alterado depois de ter sido criado
* Mesmo vazios, ocupam espaço na memória (lixos de memória)

#### Nós


* Cada item de dados é incorporado em um nó (Em um vetor, cada item de dado está em uma posição)
* Cada nó possui uma referência para o próximo nó da lista
* Um campo da própria lista contém uma referência para o primeiro nó


#### Posição X Relacionamento

* Vetor(posição)
  * Cada item ocupa uma certa posição
  * Cada posição pode ser acessada usando um número de índice

* Lista (relacionamento)
  * A única maneira de encontrar um elemento é seguir a sequência de elementos
  * Um item de dados não pode ser acessado diretamente, ou seja, o relacionamento entre eles deve ser utilizado
  * Inicia com o primeiro item, vai para o segundo, então o terceiro, até encontrar o item pesquisado
  * Pesquisa linear
  * Cada elemento da lista é armazenado em um objeto
  * Cada elemento da lista referencia o próximo e só é alocado dinamicamente quando necessário
  * Para referenciar o primeiro elemento, é utilizado um elemento chamado cabeça da lista

### Operações

* Inserir no início
* Excluir do início
* Mostrar lista
* Pesquisar
* Excluir da posição

#### Insere no início

* Insere um novo nó no início da lista
* É o local mais fácil para inserir um nó
* O proximo deste novo elemento deve ser o primeiro da lista
* A cabeça da lista aponta para o novo elemento

image.png

#### Excluir do início

* É o inverso do insere no início
* Desconecta o primeiro nó, roteando o primeiro para apontar para o segundo nó
* O segundo nó é encontrado por meio do campo **proximo** no primeiro nó
* Sobrescreve a ligação mas não diretamente o registro. Coletor de lixo linguagens de programação: verifica se existem variáveis não sendo utilizadas e limpa automaticamente (Se tratando de linguagem C, é necessário desalocar este espaço de memória pois o mesmo não possui esse automatismo).

image.png


#### Mostrar Lista e Pesquisar

* Para exibir a lista, deverá ser iniciado no primeiro elemento, seguindo a sequência de referências de nó em nó
* O final da lista é indicado pelo campo **proximo** no último nó apontando para **null/none** ao invés de outro nó
* Os nós são percorridos e é verificado se o valor do elemento é aquele que está sendo procurado (linear)
* Se atingir o final da lista sem encontrar o nó desejado, finaliza sem encontrar o elemento


#### Excluir da Posição

* O nó a ser eliminado deve ser localizado (pesquisa)
* Quando o nó atual for eliminado, deverá ser conectado o nó anterior ao seguinte

### Implementação

```python
import numpy as np
```

```python
class No:
  def __init__(self, valor):
    self.valor = valor
    self.proximo = None

  def exibe_no(self):
    print(self.valor)
```

```python
no1 = No(1)
no1.exibe_no()
```

```python
no2 = No(2)
no2.exibe_no()
```

```python
class ListaEncadeada:
  def __init__(self):
    self.primeiro = None

  def insere_inicio(self, valor):
    novo = No(valor)
    novo.proximo = self.primeiro
    self.primeiro = novo

  def exibir(self):
    if self.primeiro == None:
      print('A lista está vazia!')
      return None
    atual = self.primeiro
    while atual != None:
      atual.exibe_no()
      atual = atual.proximo

  def pesquisa(self, valor):
    if self.primeiro == None:
      print('A lista está vazia!')
      return None

    atual = self.primeiro
    while atual.valor != valor:
      if atual.proximo == None:
        return None
      else:
        atual = atual.proximo

    return atual

  def excluir_inicio(self):
    if self.primeiro == None:
      print('A lista está vazia!')
      return None

    temp = self.primeiro
    self.primeiro = self.primeiro.proximo
    return temp

  def excluir_posicao(self, valor):
    if self.primeiro == None:
      print('A lista está vazia!')
      return None

    atual = self.primeiro
    anterior = self.primeiro
    while atual.valor != valor:
      if atual.proximo == None:
        return None
      else:
        anterior = atual
        atual = atual.proximo

    if atual == self.primeiro:
      self.primeiro = self.primeiro.proximo
    else:
      anterior.proximo = atual.proximo

    return atual
```
